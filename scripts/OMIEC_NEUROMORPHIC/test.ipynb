```python
import pandas as pd

df = pd.read_excel('../../data/raw/OMIEC_NEUROMORPHIC_20_08_24.xls')
```

```python
# Importação das bibliotecas
import spacy
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np

# Carregar o modelo do spaCy (use um modelo de língua apropriado, aqui usamos en_core_web_md)
nlp = spacy.load('en_core_web_md')

# Exemplo: DataFrame com abstracts
# df = pd.DataFrame({'Abstract': ['Texto do abstract 1', 'Texto do abstract 2', ...]})

# Função para gerar embeddings com spaCy
def get_embeddings(documents):
    embeddings = []
    for doc in documents:
        # Processar o texto e pegar o vetor
        doc_nlp = nlp(doc)
        embeddings.append(doc_nlp.vector)
    return np.array(embeddings)

# Gerar os embeddings para os abstracts
embeddings = get_embeddings(df['Abstract'])

# Número de clusters que deseja criar
num_clusters = 5

# Aplicar o KMeans para clusterizar os embeddings
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(embeddings)

# Adicionar os rótulos dos clusters ao DataFrame original
df['Cluster'] = kmeans.labels_

# Mostrar os primeiros resultados
print(df.head())

# Agora, você pode explorar os clusters
print(df['Cluster'])
```
